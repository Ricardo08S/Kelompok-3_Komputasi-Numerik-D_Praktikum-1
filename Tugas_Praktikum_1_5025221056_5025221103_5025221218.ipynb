# Soal Praktikum 1
Anda sudah mengerti algoritma pemrosesan metode Bolzano, dan anda sudah memahami cara kerjanya. Sekarang anda tinggal mengimplementasikan algoritma tersebut menjadi sebuah program komputer metode Bolzano (yang dapat menampilkan proses iteratif numerik, lengkap dengan grafik fungsinya).

```
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 + x**2 - 3*x - 3

def bolzano_method(x0, x1, iterasi, toleransi=1e-3):
    print("-" * 55)
    print("\t\t     METODE BOLZANO\n")
    print("iterasi\tx0\tx1\txt\tf(x0)\tf(x1)\tf(xt)")
    print("-" * 55)

    n = 1
    root_found = False
    while n <= iterasi:
        xt = (x0 + x1) / 2
        f0 = f(x0)
        f1 = f(x1)
        ft = f(xt)
        print(f"{n}\t{x0:.3f}\t{x1:.3f}\t{xt:.3f}\t{f0:.3f}\t{f1:.3f}\t{ft:.3f}")

        if abs(ft) < toleransi:
            print("\nAkar ditemukan pada iterasi", n)
            print("Akar bernilai = ", xt)
            root_found = True
            break
        elif ft < 0:
            x0 = xt
        else:
            x1 = xt
        n += 1

    if not root_found:
        print("\nAkar belum ditemukan setelah", iterasi, "iterasi.")
        plt_title = 'Grafik Fungsi'
    else:
        plt_title = f'Grafik Fungsi (Akar: {xt})'

    print("-" * 55)

    # Plot fungsi
    x_vals = np.linspace(x0 - 1, x1 + 1, 400)  # Memperluas rentang x untuk melihat lengkungan
    y_vals = f(x_vals)
    plt.plot(x_vals, y_vals, label='f(x)')

    # Plot f(x) = 0
    plt.axhline(0, color='r', linestyle='--', label='f(x) = 0')

    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(plt_title)
    plt.grid(True)
    plt.legend()

    # Set batas x dan y untuk menampilkan lebih banyak dari grafik
    plt.xlim([x0 - 1, x1 + 1])
    plt.ylim([min(y_vals) - 1, max(y_vals) + 1])

    plt.show()

print('f(x) = x^3 + x^2 - 3x - 3')
x0 = float(input("\nMasukkan x0: "))
x1 = float(input("Masukkan x1: "))
iterasi = int(input("Jumlah iterasi: "))

bolzano_method(x0, x1, iterasi)
```
